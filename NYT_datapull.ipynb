```python
import pynytimes
from pynytimes import NYTAPI
api_key = "your_api_key"
```

```python
# Create a new NYTAPI object with your API key.
nyt = NYTAPI(api_key, parse_dates=True)
```

```python
# Fetch articles with a specific query
articles = nyt.article_search(query="Israel")
```

```python
# Number of articles pulled
len(articles)
```

```python
# See the first article
articles[0]
```

```python
# See the abstract of first article in the articles dictionary
articles[0]["abstract"]
```

```python
# See the keys in articles dictionary
articles[0].keys()
```

```python
# Fetch the top stories
top_stories = nyt.top_stories()
```

```python
# See the first top story
top_stories[0]
```

```python
# Display the keys in top stories dictionary
top_stories[0].keys()
```

```python
# Fetch the most viewed articles
nyt.most_viewed()
```
